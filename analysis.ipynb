```python
import os
import glob
import pandas as pd
import numpy as np
os.chdir("C:/Users/ab/Documents/GitHub/bundesliga_analysis/")
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
#add a "season" column to distinguish year, and combine all files in the list
combined_csv = pd.concat([pd.read_csv(f).assign(Season=os.path.basename(f).split('.')[0]) for f in all_filenames])
#export to csv
combined_csv.to_csv("data.csv", index=False, encoding='utf-8-sig')
```
